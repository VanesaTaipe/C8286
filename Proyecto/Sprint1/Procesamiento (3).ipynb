**Paso 1: Eliminar las URL**<BR>
Los localizadores uniformes de recursos (URL) son direcciones web que suelen aparecer en el texto. Por lo general, no brindan informaci√≥n significativa sobre el sentimiento y se pueden eliminar para que el texto se centre m√°s en el contenido en s√≠.


```python
import re

def remove_urls(text):
    return re.sub(r'http\S+', '', text)

data['tweet'] = data['tweet'].apply(remove_urls)
```

**Paso 2: Manejo de Emojis**

Los emojis son s√≠mbolos gr√°ficos que pueden transmitir emociones. Puedes optar por conservar, eliminar o reemplazar emojis con sus equivalentes textuales (por ejemplo, convertir üòä en "sonrisa"). La elecci√≥n depende de si desea incorporar el sentimiento emoji en su an√°lisis.

```python
import emoji
from multiprocessing import Pool

emoji_dict = {
    'üòä': 'alegre',
    'üòÇ': 'sonriente',
    '‚ù§Ô∏è': 'amor',
    'üò¢': 'triste',
    'üò≠': 'tristeza'
}

def replace_emojis(text):
    for emoji_code, emoji_text in emoji_dict.items():
        text = text.replace(emoji_code, f' {emoji_text} ')

    # Eliminar los emojis no definidos en el diccionario
    text = ''.join(char for char in text if not emoji.is_emoji(char))

    return text

def process_tweets(data):
    data['tweet_con_emoji_text'] = data['tweet'].apply(replace_emojis)
    return data

if __name__ == '__main__':

    # Dividir los datos en chunks para procesarlos en paralelo
    chunk_size = len(data) // 4  # Ajusta el n√∫mero de chunks seg√∫n tus necesidades
    chunks = [data[i:i+chunk_size] for i in range(0, len(data), chunk_size)]

    # Crear un pool de procesos
    pool = Pool()

    # Procesar los chunks en paralelo
    results = pool.map(process_tweets, chunks)

    # Combinar los resultados en un solo dataframe
    processed_data = pd.concat(results)

    # Cerrar el pool de procesos
    pool.close()
    pool.join()

```

```python
processed_data
```

En este caso necesitamos eliminar la columna **tweet**

```python
data = processed_data.drop('tweet', axis=1)
```

```python
import pandas as pd
import re

data['text_sin_'] = data['tweet_con_emoji_text'].str.replace(r'@\w+', '', regex=True)
```

```python
data['text_limpio'] = data['text_sin_'].str.replace(r'(#\w+|RT\s:)', '', regex=True)
```
